```lua
require 'hdf5'

img_size = 80
function reformat_set(data, label)
    local tot = data:size(1)
    local img_size = math.sqrt(data:size(2))
    local dat = data:resize(tot, 1, img_size, img_size)
    
    local lab = label+1
    
    local tab = {data = dat:double(), label = lab}
    function tab:size() 
        return self.data:size(1) 
    end
    
    setmetatable(tab, 
    {__index = function(t, i) 
                    return {t.data[i], t.label[i]} 
                end}
    );
    
    return tab
end

function read_set(filename, set_name)
    local myFile = hdf5.open(filename, 'r')
    local data = myFile:read(set_name .. '_x'):all()
    local label = myFile:read(set_name .. '_y'):all()
    local data_set = reformat_set(data, label)
    myFile:close()

    return data_set
end

filename = '../sample.hdf5'

all_data = {}
for i, field in pairs({'train', 'test', 'val'}) do
    all_data[field] = read_set(filename, field)
end

print(all_data)
```

```lua
function show_lab_img(t, n_rows)
    local row_size = 6
    local tot_img = t.data:size(1)
    local ind = 1
    for i = 1,n_rows do 
        local row = {}
        for n = 1,row_size do
            ind = (i-1)*row_size + n
            
            if ind > tot_img then break end
            
            local img = t.data[ind]
            img = torch.repeatTensor(img, 3, 1, 1)

            img[{{t.label[ind]}, {}, {}}] = 0

            row[n] = img
        end
        
        if table.getn(row) > 0 then
            itorch.image(row)
        end
        
        if ind > tot_img then break end
    end
end


show_lab_img(all_data['train'], 5)
```

```lua
require 'nn';

net = nn.Sequential()
net:add(nn.SpatialConvolution(1, 6, 5, 5)) -- 1 input image channel, 6 output channels, 5x5 convolution kernel
net:add(nn.ReLU())                       -- non-linearity 
net:add(nn.SpatialMaxPooling(2,2,2,2))     -- A max-pooling operation that looks at 2x2 windows and finds the max.
net:add(nn.SpatialConvolution(6, 16, 5, 5))
net:add(nn.ReLU())                       -- non-linearity 
net:add(nn.SpatialMaxPooling(2,2,2,2))
net:add(nn.View(16*17*17))                    -- reshapes from a 3D tensor of 16x5x5 into 1D tensor of 16*5*5
net:add(nn.Dropout())
net:add(nn.Linear(16*17*17, 300))             -- fully connected layer (matrix multiplication between input and weights)
net:add(nn.ReLU())                       -- non-linearity 
net:add(nn.Dropout())
net:add(nn.Linear(300, 50))
net:add(nn.ReLU())                       -- non-linearity 
net:add(nn.Dropout())
net:add(nn.Linear(50, 2))                   -- 10 is the number of outputs of the network (in this case, 10 digits)
net:add(nn.LogSoftMax())                     -- converts the output to a log-probability. Useful for classification problems

print('Lenet5\n' .. net:__tostring());

criterion = nn.ClassNLLCriterion()

trainer = nn.StochasticGradient(net, criterion)
trainer.learningRate = 0.001
trainer.maxIteration = 120 -- just do 5 epochs of training
```

```lua
-- criterion = criterion:cl()
-- net = net:cl()
-- trainset.data = trainset.data:cl()
-- trainset.label = trainset.label:cl()

trainset = all_data['train']
trainer:train(trainset)
```

```lua
testset = all_data['val']

class_performance = {0, 0}
class_n_samples = {0, 0}

bad_class_ind = {}
for i = 1,testset:size() do
    local groundtruth = testset.label[i]
    local prediction = net:forward(testset.data[i])
    local confidences, indices = torch.sort(prediction, true)  -- true means sort in descending order
    if groundtruth == indices[1] then
        class_performance[groundtruth] = class_performance[groundtruth] + 1
    else
        table.insert(bad_class_ind, i)
    end
    class_n_samples[groundtruth] = class_n_samples[groundtruth] + 1
    
end

for i, n_elem in pairs(class_n_samples) do
    print(i .. ') ' .. class_performance[i]/class_n_samples[i])
end

```

```lua
tot_ind = table.getn(bad_class_ind)
print(tot_ind)
table_reduc = {data = torch.Tensor(tot_ind, 1, 80, 80), label= torch.Tensor(tot_ind)}
for i, ind in pairs(bad_class_ind) do
    local dd = testset[ind]
    table_reduc.data[i] = dd[1]
    table_reduc.label[i] = dd[2]
end
show_lab_img(table_reduc, 100)
```
